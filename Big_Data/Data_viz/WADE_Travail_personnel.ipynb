En conclusion, si le graphe est peu chargé en couleur et contient peu d’information en terme d’occupation de l’espace, alors son entropie est faible.
Inversement, plus le graphe sera coloré et prendra la totalité de l’image (moins il y aura de fond blanc), plus l’entropie sera forte.
