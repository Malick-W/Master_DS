```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np

path = "/users/mmath/wade/Bureau/Master_DS/Data_viz/images/im1.jpeg"
img_1 = mpimg.imread(path)

plt.imshow(img_1)
plt.show()
```

```python
def rgb2gray(rgb):
    """ convertion d'image en gris """
    return np.dot(rgb[...,:3], [0.299, 0.587, 0.144])


img_1_gris = rgb2gray(img_1)
plt.imshow(img_1_gris, cmap = plt.get_cmap('gray'))

path = "/users/mmath/wade/Bureau/Master_DS/Data_viz/images/image1_gris.jpg"

plt.savefig(path)
plt.show()
```
